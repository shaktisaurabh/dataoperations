```python
from nltk.corpus import stopwords 
def remove_stopwords(amn):
    L=[] 
    for text in amn:
        if text not in stopwords.words('english'):
            L.append(text)
    return L
#stopwords is important for sentence formation but add no semantic value to the sentence and hence can 
#be done away with
```

```python
import re
a1='<p>This is dhoni</p><br><img>'
am1=re.findall(r'<.*>',a1) 
print(am1) 
#this pattern r'<.*>' cannot be used to remove all the html tags as the pattern r'<.*>' deals with all
#the characters between the first opening tag and the last closing tag. Hence it might end up reducing the 
#entire string as empty if it is replaced. Thus this isnot right thing to do. 
```

```python
import re 
a1='<p>When will he come</p><br><p>Who is he</p><br><img>'
am1=re.findall(r'<.*>',a1) 
print(am1)  
#As you can see r'<.*>' this regex extracts the entire text between the first opening bracket and the
#last closing bracket. It is not right and thus we use this pattern r'<.*?>' and this pattern only takes
#all characters between first opening and first closing bracket and so on. 
```

```python
import re 
a1='<p>This is dhoni</p><br><img>' 
am1=re.findall(r'<.*?>',a1) 
print(am1) 
#r'<.*?> this pattern captures everything between first occurence of < and first occurence of > and hence
#we get 4 items in the list <p>,</p>,<br> and <img>. This regex pattern r'<.*?>' can exclude all the
#html tags from the data. 
```
